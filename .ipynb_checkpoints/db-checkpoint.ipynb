```python
import os
from dotenv import load_dotenv
import psycopg2
```

```python
load_dotenv()

DB_NAME = os.getenv("DB_NAME")
DB_USER = os.getenv("DB_USER")
DB_HOST = os.getenv("DB_HOST")
```

```python
def get_conn():
    return psycopg2.connect(
        dbname = DB_NAME,
        user = DB_USER,
        host = DB_HOST
    )
```

```python
if __name__ = "__main__":
    conn = get_conn()
    print("Connected")
    conn.close()
```
