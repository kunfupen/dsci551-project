```python
import os
from dotenv import load_dotenv
import psycopg2
```

```python
load_dotenv()

DB_NAME = os.getenv("DB_NAME")
DB_USER = os.getenv("DB_USER")
DB_HOST = os.getenv("DB_HOST")
```

```python
def get_conn():
    return psycopg2.connect(
        dbname = DB_NAME,
        user = DB_USER,
        host = DB_HOST
    )
```

```python
if __name__ == "__main__":
    conn = get_conn()
    print("Connected")
    conn.close()
```

```python
def search_by_author(author):
    conn = get_conn()
    cur = conn.cursor()
    cur.execute("""
        SELECT title, authors, average_rating
        FROM books
        WHERE authors ILIKE %s
    """, (f"%{author}%",))
    results = cur.fetchall()
    cur.close()
    conn.close()
    return results

def filter_by_num_pages(min_pages, max_pages):
    conn = get_conn()
    cur = conn.cursor()
    cur.execute("""
        SELECT title, authors, num_pages
        FROM books
        where num_pages BETWEEN %s AND %s
        """, (min_pages, max_pages))
    results = cur.fetchall()
    cur.close()
    conn.close()
    return results

def top_rated_books(limit):
    conn = get_conn()
    cur = conn.cursor()
    cur.execute("""
        SELECT title, authors, average_rating
        FROM books
        ORDER BY average_rating DESC
        LIMIT %s
    """, (limit,))
    results = cur.fetchall()
    cur.close()
    conn.close()
    return results
        
```
